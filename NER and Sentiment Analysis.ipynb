```python
import re
import string
from nltk.corpus import stopwords
english_stopwords = stopwords.words('english')

def remove_html(text):
    html = re.compile(r"<,*?>")
    return html.sub(r"",text)

def remove_URL(text):
    url = re.compile(r"https?://\S+|www\.\S+")
    return url.sub(r"", text)

def remove_emoji(string):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  #emoticons
        u"\U0001F300-\U0001F5FF"  #symbols & pictographs
        u"\U0001F680-\U0001F6FF"  #transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  #flags (IOS)
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+",
        flags = re.UNICODE,
    )
    return emoji_pattern.sub(r"", string)

def remove_punct(text):
    table = str.maketrans("", "", string.punctuation)
    return text.translate(table)

def remove_stopwords(text):
    text = [word.lower() for word in text.split() if word.lower() not in english_stopwords]
    return " ".join(text)
```

```python
data["clean_text"] = data["text"]
data["clean_text"] = data.clean_text.map(lambda x: remove_html(x))
data["clean_text"] = data.clean_text.map(lambda x: remove_URL(x))
data["clean_text"] = data.clean_text.map(lambda x: remove_emoji(x))
data["clean_text"] = data.clean_text.map(lambda x: remove_punct(x))
data["clean_text"] = data.clean_text.map(lambda x: remove_stopwords(x))
data.head(2)
```

```python
import spacy
from spacy import displacy
spacy.prefer_gpu()
nlp = spacy.load("en_core_web_trf")

ModelPATH = "C:/Users/hadia/Desktop/GOALS/VeeSmart internship/Spacy/model/BERTmodel"
```

```python
from transformers import DistilBertTokenizerFast
tokenizer = DistilBertTokenizerFast.from_pretrained("bert-base-uncased")
from transformers import TFDistilBertForSequenceClassification
model = TFDistilBertForSequenceClassification.from_pretrained(ModelPATH)
import tensorflow_addons as tfa
optimizer = tfa.optimizers.AdamW(weight_decay=1e-4, learning_rate=5e-5, epsilon = 1e-09)
loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
metric = tf.keras.metrics.SparseCategoricalAccuracy('accuracy')
model.compile(optimizer=optimizer, loss=loss, metrics=metric)
```

```python
import numpy as np
def analyse(x):
    global model
    
    review = x['text']
    date = str(x['date'])
    labels = [["neg"],["pos"]]
    if x['stars']>3:
        WholeReviewSentiment = labels[1]
    else:
        WholeReviewSentiment = labels[0]
    
    opinionMining=[]
    doc=nlp(review)
    for i in list(doc.sents):
        aspects=set()
        predict_input = tokenizer.encode(str(i), max_length=262, truncation=True, padding="max_length", return_tensors="tf")
        tf_output = model.predict(predict_input)[0]
        tf_prediction = tf.nn.softmax(tf_output, axis=1)
        label = tf.argmax(tf_prediction, axis=1).numpy()
        sent=labels[label[0]]
        for token in i:
            if token.pos_ == "NOUN":
                aspects.add(token.lemma_)
        if len(aspects)==0:
            aspects=""
        
        opinionMining.append({"SubReview":str(i), "aspects":aspects, "sentiment":sent})        
                
    show={"Review":review, "Date":date, "opinionMining":opinionMining, "WholeReviewSentiment":WholeReviewSentiment}
    return show
```

```python
for index, row in data.iterrows():
    print(analyse(row))
    print("_"*125,"\n","_"*125)
```
